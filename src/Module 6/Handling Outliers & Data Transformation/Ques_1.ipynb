```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example dataset (you can replace this with your actual data)
data = [10, 12, 13, 11, 12, 14, 110, 12, 13, 11]  # Outlier: 110
df = pd.DataFrame({'Values': data})

# Create the boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Values'], color='lightgreen')

# Add labels and title
plt.title("Boxplot for Visualizing Outliers")
plt.xlabel("Values")
plt.grid(True)

# Show plot
plt.show()
```
